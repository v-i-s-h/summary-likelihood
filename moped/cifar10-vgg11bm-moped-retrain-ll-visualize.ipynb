To visualize the results of retraining

```python
import pickle

import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
with open('retrain_logs.pkl', 'rb') as f:
    logs = pickle.load(f)
```

```python
train_log = logs['train_log']
retrain_log = logs['retrain_log']
```

```python
fig, ax = plt.subplots(1, 4, figsize=(4*8, 4))

n = len(train_log['loss'])

ax[0].plot(range(1, n+1), train_log['loss'])
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss')
ax[0].set_title('Loss')

ax[1].plot(range(1, n+1), train_log['ece'])
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('ECE')
ax[1].set_title('ECE')

ax[2].plot(range(1, n+1), train_log['mce'])
ax[2].set_xlabel('Epochs')
ax[2].set_ylabel('MCE')
ax[2].set_title('MCE')

ax[3].plot(range(1, n+1), train_log['err'])
ax[3].set_xlabel('Epochs')
ax[3].set_ylabel('Error')
ax[3].set_title('Error')

plt.suptitle("MOPOED training")
```

```python
fig, ax = plt.subplots(1, 4, figsize=(4*8, 4))

n = len(retrain_log['loss'])

ax[0].plot(range(1, n+1), retrain_log['loss'])
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss')
ax[0].set_title('Loss')

ax[1].plot(range(1, n+1), retrain_log['ece'])
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('ECE')
ax[1].set_title('ECE')

ax[2].plot(range(1, n+1), retrain_log['mce'])
ax[2].set_xlabel('Epochs')
ax[2].set_ylabel('MCE')
ax[2].set_title('MCE')

ax[3].plot(range(1, n+1), retrain_log['err'])
ax[3].set_xlabel('Epochs')
ax[3].set_ylabel('Error')
ax[3].set_title('Error')

plt.suptitle("Retraining without prior KL")
```

```python
fig = plt.figure(figsize=(8, 6))
plt.plot(range(1, n+1), train_log['ece'])
plt.xlabel('Epochs')
plt.ylabel('ECE')
plt.title('MOPED training with ELBO')
plt.plot([1, 25], [0.092, 0.108], ls=':')
```

```python
fig = plt.figure(figsize=(8, 6))
plt.plot(range(1, n+1), retrain_log['ece'])
plt.xlabel('Epochs')
plt.ylabel('ECE')
plt.title('Retraining last last layer without KL loss')
plt.plot([1, 25], [0.108, 0.088], ls=':')
```
